```python
import pandas as pd
```

**SAMPLE - 1**

```python
# IMPORTING RAW DATA TO DATAFRAME

df = pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/Underweight-knn.csv")
df
```

```python
df.info()
```

```python
# INPUTS
x = df.iloc[:, 0:2].values

# OUTPUTS
y = df.iloc[:,-1].values
```

```python
# IMPORTING KNN ALGORITHM TO MODEL
# n_neighbours = 3 ==> no.of data points to be picked in the region
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
```

```python
# FITIING THE MODEL
model.fit(x,y)
```

```python
# PREDICTING THE MODEL

y_pred = model.predict(x)
y_pred
```

```python
y
```

```python
# ACCURACY SCORE

from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y,y_pred)
```

```python
# CONFUSION MATRIX b/w ACTUAL & PREDICTING OUTPUTS

confusion_matrix(y,y_pred)
```

**SAMPLE - 2**

```python
df_2 = pd.read_csv('/content/Iris.csv')
df_2
```

```python
df_2.info()
```

```python
df_2.shape
```

```python
# INPUT
x = df_2.iloc[:,1:5].values

# OUTPUT
y = df_2.iloc[:,-1].values
```

```python
# IMPORTING KNN ALGORITHM TO MODEL

from sklearn.neighbors import KNeighborsClassifier
model_1 = KNeighborsClassifier(n_neighbors=15)
```

```python
# FITTING THE MACHINE

model_1.fit(x,y)
```

```python
# PREDICTING SINGLE OUTPUT from ENTIRE MODEL

y_pred = model_1.predict([[4.7,3.2,1.3,0.21]])
y_pred
```

```python
# PREDICTING ALL OUTPUTS

y_pred = model_1.predict(x)
y_pred
```

```python
model_1.score(x,y) 
```

```python
# ACCURACY SCORE

from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

```python
# CONFUSION MATRIX

from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y,y_pred)
```

```python
print(classification_report(y,y_pred))
```
